```python
import nltk
nltk.download("all")
```

```python
from nltk.tokenize import word_tokenize

text = "I love playing table tennis"
tokens = word_tokenize(text)
print(tokens)

#character level
char_tokens = list(text)
print(char_tokens)
```

```python
text = text.lower()
```

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
print(stop_words)

print("After stopword little")
filtered_tokens = [token for token in tokens if token not in stop_words]
print(filtered_tokens)
```

```python
#Removing punctuation
import string
text = "Hello!!! I need a cup ....of tea"
clean_text = text.translate(str.maketrans('', '', string.punctuation))
print(clean_text)
```

```python
from nltk.tokenize import RegexpTokenizer
tokenizer = RegexpTokenizer(r'\w+')
text = "This is another  #45 sample Text!! #$"
clean_text = tokenizer.tokenize(text)
print(clean_text)
```

```python
from nltk.tag import pos_tag
sentence = " I Love Learning NLP"
words = nltk.word_tokenize(sentence)
print(words)

pos_tags = pos_tag(words)
print(pos_tags)
```

```python
from nltk.stem import PorterStemmer
porter_stemmer = PorterStemmer()
words = ["Running","Jumps","liked","Tunnig","playing","Happily"]
stemmed_words = [porter_stemmer.stem(word) for word in words]
print("Original Words",words)
print("Stemmed words",stemmed_words)
```

```python
#Snowball Stemmer : is an enhanced version of the Porter stemmer, supports multiple languages.

from nltk.stem import SnowballStemmer
snowball_stemmer = SnowballStemmer(language='english')
words_to_stem = ["running","jumped","happily","quickly","foxes"]
stemmed_words = [snowball_stemmer.stem(word) for word in words_to_stem]
print("Original words:",words_to_stem)
print("Stemmed words:",stemmed_words)
```

```python
#Lancaster stemmer : is known for being more aggressive and faster than other stemmers.

from nltk.stem import LancasterStemmer
lancaster_stemmer = LancasterStemmer()
words_to_stem = ["running","jumped","happily","quickly","foxes"]
stemmed_words = [lancaster_stemmer.stem(word) for word in words_to_stem]
print("Original words:",words_to_stem)
print("Stemmed words:",stemmed_words)
```
